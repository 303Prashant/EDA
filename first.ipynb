```python
import os
import pandas as pd

df  = pd.read_csv(path)
df.head(10)
```

```python
print("Dataset shape:", df.shape)
print("Heading names:\n", df.columns)
print("Missing values:\n", df.isnull().sum())
```

```python
df.head()
```

```python
newdf = df.copy()
```

```python
newdf=newdf.drop(columns=['Source', 'Official/Unofficial'])
```

```python
newdf['VEP Turnout Rate'] = newdf['VEP Turnout Rate'].str.replace('%', '').astype(float)
```

```python
newdf['Overseas Eligible'] = newdf['Overseas Eligible'].fillna(0)
```

```python
newdf['likely_to_vote'] = (newdf['VEP Turnout Rate'] >= 64).astype(int)
```

```python
newdf = newdf.dropna(subset=['State Abv'])
newdf['Vote for Highest Office (President)'] = newdf['Vote for Highest Office (President)'].fillna(0)
```

```python
newdf['% Non-citizen'] = newdf['% Non-citizen'].str.replace('%', '').astype(float)
```

```python
clean = ['Vote for Highest Office (President)','Voting-Eligible Population (VEP)','Voting-Age Population (VAP)','Total Ineligible Felon','Prison','Probation','Parole', 'Total Ballots Counted (Estimate)']

for c in clean:
    newdf[c] = newdf[c].astype(str).str.replace(',', '').replace('nan', '0').astype(float)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(14, 6))
state = newdf.sort_values('VEP Turnout Rate', ascending=False)

sns.barplot(x='State Abv', y='VEP Turnout Rate', data=state, palette='viridis')
plt.xticks(rotation=90)
plt.title('VEP Turnout Rate by State')
plt.ylabel('Turnout Rate (%)')
plt.xlabel('State')
plt.tight_layout()
plt.show()
```

```python
newdf.head(4)
```

```python
sns.countplot(x='likely_to_vote', data=newdf, palette='Set2')
plt.xticks([0, 1], ['Unlikely to Vote (<62%)', 'Likely to Vote (62%)'])
plt.title('Distribution of Turnout Classes')
plt.ylabel('Number of States')
plt.xlabel('Voter Turnout Class')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
x = newdf.drop(columns=['VEP Turnout Rate', 'likely_to_vote', 'State', 'State Abv'])
y = newdf['likely_to_vote']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# x_train
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(x_train_scaled, y_train)
```

```python
y_pred = model.predict(x_test_scaled)

# Evaluation
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
feature_importance = pd.Series(model.feature_importances_, index=x.columns)
top_features = feature_importance.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_features.values, y=top_features.index)
plt.title('Top Feature Importances')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```
